```python
# cell 2 
# 1.1 加载站点表 + 筛 Houston 工业区范围（Ship Channel）

import numpy as np

sites_path = os.path.join(RAW_DIR, files["sites"])
sites = pd.read_csv(sites_path)

print("sites shape:", sites.shape)
print("sites columns:", list(sites.columns))
display(sites.head(3))
```

```python
# Cell 3 只筛 Texas（先别急着算距离）
# 2 只筛 Texas 站点（不涉及地理计算）

sites_tx = sites[sites["State Name"] == "Texas"].copy()

print("Texas sites shape:", sites_tx.shape)
display(sites_tx[["State Name", "County Name", "City Name"]].head(10))
```

```python
# Cell 4 再筛 Houston 周边 4 个县（工业带核心）
# 3 Houston 工业区核心县

target_counties = {"Harris", "Galveston", "Chambers", "Brazoria"}

sites_hou = sites_tx[sites_tx["County Name"].isin(target_counties)].copy()

print("Houston-area sites shape:", sites_hou.shape)
display(
    sites_hou[
        ["County Name", "City Name", "Land Use", "Location Setting"]
    ].head(20)
)
```

```python
# Cell 5：最后一刀 —— 距离 Ship Channel（核心工业区）

from math import radians, sin, cos, sqrt, atan2

def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c

# Houston Ship Channel 核心参考点（固定）
center_lat, center_lon = 29.74, -95.20
radius_km = 30

sites_hou = sites_hou.dropna(subset=["Latitude", "Longitude"]).copy()

sites_hou["dist_km"] = sites_hou.apply(
    lambda r: haversine_km(center_lat, center_lon, r["Latitude"], r["Longitude"]),
    axis=1
)

sites_industrial_zone = sites_hou[sites_hou["dist_km"] <= radius_km].copy()
sites_industrial_zone = sites_industrial_zone.sort_values("dist_km")

print("Industrial zone sites:", sites_industrial_zone.shape)
display(
    sites_industrial_zone[
        ["County Name", "City Name", "Land Use", "Location Setting", "dist_km"]
    ].head(30)
)
```

```python
# Cell 6 生成站点 ID（非常重要）

# 6 生成 AQS 标准站点唯一 ID（后面 join 用）

sites_industrial_zone["site_id"] = (
    sites_industrial_zone["State Code"].astype(str).str.zfill(2)
    + "-"
    + sites_industrial_zone["County Code"].astype(str).str.zfill(3)
    + "-"
    + sites_industrial_zone["Site Number"].astype(str).str.zfill(4)
)

display(
    sites_industrial_zone[
        ["site_id", "County Name", "City Name", "Land Use", "dist_km"]
    ].head(10)
)

print("unique site_id:", sites_industrial_zone["site_id"].nunique())
```

```python
# 读取 hourly 大文件：只读需要的列 + 只保留目标站点
# 我们先做一个函数：只读关键列，然后筛站点。

def load_hourly_filtered(file_key, usecols=None):
    path = os.path.join(RAW_DIR, files[file_key])

    # 如果你不确定列名，先不传 usecols，让它读 5 行看列名
    sample = pd.read_csv(path, nrows=5)
    cols = list(sample.columns)
    print(file_key, "columns:", cols)

    # 常见关键列名（AQS hourly 通常是这些）
    needed = [
        "State Code", "County Code", "Site Num",
        "Date Local", "Time Local",
        "Sample Measurement",
        "Units of Measure",
        "Parameter Name"
    ]

    # 有些版本叫 Site Number 而不是 Site Num
    # 所以我们做一个映射
    col_map = {}
    if "Site Num" in cols:
        col_map["site_col"] = "Site Num"
    elif "Site Number" in cols:
        col_map["site_col"] = "Site Number"
    else:
        raise ValueError("找不到站点列：Site Num / Site Number")

    # 只保留存在的列
    needed_exist = [c for c in needed if c in cols]
    print("using cols:", needed_exist)

    df = pd.read_csv(path, usecols=needed_exist)

    # 标准化 site_id
    df["State Code"] = df["State Code"].astype(str).str.zfill(2)
    df["County Code"] = df["County Code"].astype(str).str.zfill(3)
    df[col_map["site_col"]] = df[col_map["site_col"]].astype(str).str.zfill(4)

    df["site_id"] = df["State Code"] + "-" + df["County Code"] + "-" + df[col_map["site_col"]]

    # 过滤目标站点
    df = df[df["site_id"].isin(target_site_ids)].copy()

    # 组合时间戳
    df["datetime_local"] = pd.to_datetime(df["Date Local"] + " " + df["Time Local"], errors="coerce")

    # 转成数值
    df["value"] = pd.to_numeric(df["Sample Measurement"], errors="coerce")

    # 清理
    df = df.dropna(subset=["datetime_local", "value"])
    return df

pm25 = load_hourly_filtered("PM25")
print(pm25.shape)
display(pm25.head(5))
```

```python
# 4 做异常审计（第一版：Spike + Jump）

def audit_anomalies(df, pollutant_name="PM2.5"):
    # df 必须包含：site_id, datetime_local, value
    out = df.copy()
    out = out.sort_values(["site_id", "datetime_local"])

    # 每站点计算分位数阈值
    q99 = out.groupby("site_id")["value"].quantile(0.99).rename("q99")
    out = out.join(q99, on="site_id")

    # Spike
    out["is_spike"] = out["value"] >= out["q99"]

    # Jump（相邻小时差分）
    out["prev_value"] = out.groupby("site_id")["value"].shift(1)
    out["diff"] = (out["value"] - out["prev_value"]).abs()

    # Jump阈值：每站点 diff 的 99分位
    d99 = out.groupby("site_id")["diff"].quantile(0.99).rename("d99")
    out = out.join(d99, on="site_id")
    out["is_jump"] = out["diff"] >= out["d99"]

    # 输出异常事件
    events = out[out["is_spike"] | out["is_jump"]].copy()
    events["pollutant"] = pollutant_name

    # 异常评分（简单粗暴但有效）
    events["score"] = 0
    events.loc[events["is_spike"], "score"] += 1
    events.loc[events["is_jump"], "score"] += 1
    events["score"] += (events["value"] / (events["q99"] + 1e-9))  # 相对强度

    events = events.sort_values(["score"], ascending=False)
    return events

pm25_events = audit_anomalies(pm25, "PM2.5")
print("pm25_events:", pm25_events.shape)
display(pm25_events.head(30)[["site_id","datetime_local","value","q99","diff","d99","is_spike","is_jump","score"]])

```
